```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
dt1=pd.read_csv('/content/fraudTest.csv')
dt1.head()
```

```python
dt2=pd.read_csv('/content/fraudTrain.csv')
dt2.head()
```

concat the two large dataset

```python
df_num=dt1.select_dtypes(np.number)
```

```python
df_num.info()
```

CHECKING MISSING VALUES

```python
df_num.notnull().sum()
```

```python
df_num['is_fraud'].value_counts()
```

the dataset is highly unbalance

fraud transaction-------> 1
legit transaction-------> 0

```python
legit=df_num[df_num.is_fraud==0]
fraud=df_num[df_num.is_fraud==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.amt.describe()
```

```python
fraud.amt.describe()
```

```python
 legit_sample=legit.sample(n=2145)
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['is_fraud'].value_counts()
```

```python
new_dataset.groupby('is_fraud').mean
```

spliting the data

```python
X = new_dataset.drop(columns='is_fraud', axis=1)
Y = new_dataset['is_fraud']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
new_dataset.columns
```

model training Random forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier()
```

```python
model.fit(X_train,Y_train)
```

accuracy score

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```

getting the more accurracy that is 94.1% on test data
